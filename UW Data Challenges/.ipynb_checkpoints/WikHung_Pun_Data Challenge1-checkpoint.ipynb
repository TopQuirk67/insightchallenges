# Preprocess data
* Fill in missing values
* Create train/test sets

```python
seed = 2018
```

```python
# Naive method: Fill in missing values with category mean
df['nuclei'] = df['nuclei'].fillna(df['nuclei'].mean())
```

```python
#df['mitoses'] = df['mitoses'].apply(lambda x: 1 if x == 1 else 0)
```

```python
# Descriptive statistics after transformation / feature engineering
```

```python
# Split the features and label
X = df[[c for c in df.columns if c not in ['class', 'id']]].values
y = df['class'].values
```

# Baseline Model

With logistic regression as the baseline model, I used 10-fold stratified cross-validation to estimate the model performance.

```python
KFold = 10
```

```python
clf = LogisticRegression(penalty='l1', random_state=seed)
```

```python
print('Stratified {}-Fold Cross-validation'.format(KFold))
print('Logistic regression (L1-regularization)')
precisions, recalls, fscores, aucs, accs = model_eval(clf, X, y, KFold, seed)

print('Precision: {:.2f}'.format(np.mean(precisions)))
print('Recalls: {:.2f}'.format(np.mean(recalls)))
print('F1-score: {:.2f}'.format(np.mean(fscores)))
print('AUC: {:.2f}'.format(np.mean(aucs)))
print('Accuracy: {:.2f}'.format(np.mean(accs)))
```
