How many numbers under a 1000 have three distinct integer factors?

```python
# brute force
n = set()
m = set()
nn = 0
for i in range(1,501):
    for j in range(1,501):
        for k in range(1,501):
            if (i*j*k<1000) and (i!=j) and (i!=k) and (j!=k):
                nn += 1
                print(nn,tuple(sorted([i,j,k])),i*j*k)
                n.add(tuple(sorted([i,j,k])))
                m.add(i*j*k)
```
